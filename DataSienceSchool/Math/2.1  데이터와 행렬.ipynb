```python
from sklearn.datasets import load_digits

digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31]
d = []
for i in range(8):
    d.append(digits.images[samples[i]]) # 1차원 벡터 8*8

plt.figure(figsize=(8, 2))
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("image {}".format(i + 1))
plt.suptitle("Images of 0's and 1's")
plt.tight_layout()
plt.show()
```

```python
digits.images[2]
```

```python
v = []
for i in range(8):
    v.append(d[i].reshape(64, 1))
    
plt.figure(figsize=(8, 3)) #  64-크기의 1차원 벡터 - 같은 숫자에 대한 벡터가 서로 닮음
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(v[i], aspect=0.4,
               interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False); plt.xticks([]); plt.yticks([])
    plt.title("vector {}".format(i + 1))
plt.suptitle("vectorized images", y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

### 행렬
> 복수의 차원을 가지는 데이터 레코드가 여러개 있는 경우의 데이터를 합쳐서 포기한 것

벡터는 열의 수가 1인 행렬이라고 볼 수 있음 -> 열 벡터라고도 함. (column vector)

하나의 데이터 레코드를 단독으로 벡터로 나타낼 때는 하나의 열(column)나타냄 (세로로 하나를 나타낸다는 것)   
복수의 데이터 레코드 집합을 행렬로 나타낼 때는 하나의 데이터 레코드가 하나의 행(row)으로 나타냄 

```python
x1 = np.array([[5.1], [3.5], [1.4], [0.2]]) # 하나의 데이터 레코드를 단독으로 벡터로 나타낸 것
x1
```

스칼라와 벡터도 수학적으로 행렬에 속함  
스칼라 : 열, 행의 수가 각각 1인 행렬
벡터 : 열의 수가 1인 행렬

### 전치 연산

스칼라, 벡터, 행렬 데이터를 변형시키는 방법 
> 전치 연산이란 행렬의 행과 열을 바꾸는 연산

열 백터 전치 연산 -> 행의 수가 1인 행렬 -> 행 벡터

```python
A
```

### 특수한 벡터와 행렬

#### 영벡터 zeros-vector
모든 원소가 0인 N차원 벡터 : 영벡터

```python
np.zeros((3,1)) # row * column 
```

#### 일벡터 ones-vector

모든 원소가 1인  N차원 벡터 

```python
np.ones((3, 1))
```

#### 정방행렬 
행의 개수와 열의 개수가 같은 행렬

#### 대각행렬
대각 : 행렬에서 행과 열이 같은 위치 (주대각)  
비대각 : 대각 위치에 있지 않은 것들   
모든 비대각 요소가 0인 행렬
- 꼭 정방행렬이 아니여도 됨

```python
np.diag([1,2,3]) # 대각 행렬 생성
```

```python
np.diag([1,2,3,4,5])
```

#### 항등행렬
대각행렬 중에서도 모든 대각성분의 값이 1인 대각행렬. 보통 I로 표시함.

```python
np.identity(3)
```

```python
np.eye(4)
```

#### 대칭행렬
전치 연산을 통해 얻은 전치행렬과 원래 행렬이 같으면 대칭행렬이라고 함.
정방행렬만 대칭행렬이 될 수 있음

```python
a = np.array([['개' ,'똥','아'], ['똥','싸','니'], ['아','니','요']])
```
