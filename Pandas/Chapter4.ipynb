```python
import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill']) # seaborn을 활용해서 그래프그리기  (히스토그램)
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False) # 밀집도 그래프 제외 (주어진 데이터를 정규화시켜 넓이가 1이 되도록 그린 그래프)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 밀집도 그래프만 나타내기
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Den')
```

```python
# 양탄자 그래프
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True) # 양탄자 그래프 -> 그래프의  축에 동일한 길이의 직선을 붙여 데이터의 밀접 정도를 표현함
ax.set_title('Total Bill Histogram width Density and Rug plot')
ax.set_xlabel('Total Bill')
```

### count 그래프


```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

### 산점도 그래프

```python
ax = plt.subplots()
```

```python
ax = sns.regplot(x='total_bill', y='tip', data=tips) # 산점도 그래프
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
```

####  산점도 그래프, 히스토그램 같이 보여주기 Jointplot

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips,)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') # 육각형 산점도
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

### 이차원 밀집도 

```python
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'], data2=tips['tip'], shade=True) # shade : 그래프에 음영주기
```

### 바 그래프
지정한 변수의 평균을 계산하여 그릴 수 있음

```python
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

### 박스 그래프 
최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 다양한 통계랑을 한번에 표현함

```python
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```
