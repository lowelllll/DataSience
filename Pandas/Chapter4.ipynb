## 데이터 시각화
그래프 그리기

```python
import seaborn as sns
```

```python
anscombe = sns.load_dataset('anscombe') # 앤스콤 데이터 집합
print(anscombe)
print(type(anscombe))
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
```

```python
plt.plot(dataset_1['x'], dataset_1['y']) # 선그래프
```

```python
plt.plot(dataset_1['x'], dataset_1['y'], 'o') # 점그래프
```

### 앤스콤 데이터 집합 그래프 그리기

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
fig = plt.figure() # 격자가 위치할 기본틀 생성
```

```python
axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4) # 그래프 격자를 그림 (그래프 행 크기, 그래프 열 크기)
```

```python
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')
```

```python
axes1.set_title('dataset_1') #제목달기
axes2.set_title('dataset_2') 
axes3.set_title('dataset_3') 
axes4.set_title('dataset_4') 
```

```python
fig.suptitle('Anscombe Data') # 기본 틀에 제목 추가
```

```python
fig.tight_layout() # 레이아웃 조절
```
