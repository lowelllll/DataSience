##  데이터 처리! 

```python
sc['Born'].dtype
```

```python
born_datetime = pd.to_datetime(sc['Born'], format='%Y-%m-%d')
born_datetime
```

```python
died_datetime = pd.to_datetime(sc['Died'], format='%Y-%m-%d')
died_datetime
```

```python
sc['born_dt'], sc['died_dt'] = (born_datetime, died_datetime)
```

```python
sc
```

```python
sc.shape
```

```python
sc['age_days_dt'] = (sc['died_dt'] - sc['born_dt']) # 날짜 계산 
sc
```
