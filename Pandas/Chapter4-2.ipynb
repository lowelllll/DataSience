```python
tips = sns.load_dataset('tips')
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot) # 대각선 기준으로 위쪽에 그릴 그래프 지정
pair_grid = pair_grid.map_lower(sns.kdeplot) # 대삭선 기준으로 아래쪽에 그릴 그림 지정
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) # 대각선을 중심으로 그래프를 그림
plt.show()
```

```python
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
```

### FaceetGrid 

```python
facet = sns.FacetGrid(tips, col='time') # 구분할 열 이름을 전달해서 객체를 생성함
facet.map(sns.distplot, 'total_bill', rug=True) # 그래프의 종류와 x축으로 지정할 열 이름을 순서대로 전달.
```

```python
facet = sns.FacetGrid(tips, col='day', hue='sex') # tips 데이터 집합의 day열로 그룹을 구분해 그래프를 그린 것.
facet = facet.map(plt.scatter, 'total_bill', 'tip') # 산점도 그래프를 그리기 위해 scatter를 전달함.
facet = facet.add_legend()
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```
