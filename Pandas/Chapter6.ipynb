### 누락 값 처리하기!

```python
ebola.fillna(0).iloc[0:10, 0:5] # 행, 열 fillna 메서드에 0을 대입하면 누락값을 0으로 변경함.
```

```python
ebola.fillna(method='ffill').iloc[0:10, 0:5] # 누락값이 나타나기 전의 값으로 변경. (그 전 행 값을 사용함)
```

```python
ebola.fillna(method='bfill').iloc[0:10, 0:5] #  누락값이 나타난 이후의 첫번째 값으로 변경
```

```python
ebola.interpolate().iloc[0:10, 0:5] # 누락값 양쪽의 값을 이용해 중간값으로 처리함
```

```python
ebola.iloc[0:10, 0:5]
```
